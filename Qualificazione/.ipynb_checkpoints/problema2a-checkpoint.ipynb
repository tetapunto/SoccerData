### Import dataset

```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


data = pd.read_json('worldCup.json')
players = pd.read_json('player.json')

data.head()
```

### trasformo colonna tags da lista di dict a lista 

```python
def make_list(list_dict, key = 'id'):
    if len(list_dict) ==0:
        return []
    return [i['id'] for i in list_dict]
    
    
data.tags = data.tags.apply(make_list)
data.head()
```

### Conto i passaggi di ogni giocatore

```python
pass_for_player = data[data.eventId ==8].groupby(['playerId','teamId']).size().reset_index(name = 'total_pass')
pass_for_player
```

### Seleziono prima tutti gli eventi accurati e poi scelgo quelli che sono anche passaggi

```python
accurate_events = data.loc[[idx for idx,row in data.iterrows() if 1801 in row['tags']]]
accurate_pass_for_player = accurate_events[accurate_events.eventId ==8].groupby(['playerId','teamId']).size().reset_index(name = 'accurate_pass')
accurate_pass_for_player
```

### Calcolo accuratezza + media per squadra + dev per squadra  

```python
pass_summary = pass_for_player.merge(accurate_pass_for_player, on = ['playerId','teamId'])
pass_summary['accuracy'] =pass_summary.accurate_pass/pass_summary.total_pass
```

```python
mean_grp = pass_summary.groupby('teamId')['accuracy'].mean().reset_index(name = 'mean_accuracy_by_team')
```

```python
dev_grp = pass_summary.groupby('teamId')['accuracy'].std().reset_index(name = 'std_accuracy_by_team')
```

### Aggiusto formato risultati

```python
x = mean_grp.merge(dev_grp, on =['teamId'])
x.columns = ['nome_squadra','accuratezza_media_dei_passaggi','standard_deviation_accuratezza_media_dei_passaggi']
```

```python
x.round({'accuratezza_media_dei_passaggi':2,'standard_deviation_accuratezza_media_dei_passaggi':2})
```
