### Import dataset

```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


data = pd.read_json('worldCup.json')
players = pd.read_json('player.json')

data.head()
```

### trasformo colonna tags da lista di dict a lista 

```python
def make_list(list_dict, key = 'id'):
    if len(list_dict) ==0:
        return []
    return [i['id'] for i in list_dict]
    
    
data.tags = data.tags.apply(make_list)
data.head()
```

## Divido il campo in celle

```python
# dimensioni stadio Luzhniki
lungo = 68
largo = 105


# griglia
# 1,5 ..... 5,5
# .........
# 1,1 .... 5,1

def give_cell(pos):
    
    if pos['x'] < lungo/5:
        xc = 1
    elif pos['x'] < (lungo/5)*2:
        xc = 2
    elif pos['x'] < (lungo/5)*3:
        xc = 3
    elif pos['x'] < (lungo/5)*4:
        xc = 4
    else:
        xc = 5

    if pos['y'] < largo/5:
        yc = 1
    elif pos['y'] < (largo/5)*2:
        yc = 2
    elif pos['y'] < (largo/5)*3:
        yc = 3
    elif pos['y'] < (largo/5)*4:
        yc = 4
    else:
        yc = 5
            
        
    return [xc,yc]
    
def get_grid(lis):
    if lis[0] == 1:
        if lis[1] == 1:
            return 1
        elif lis[1] == 2:
            return 6
        elif lis[1] == 3:
            return 11
        elif lis[1] == 4:
            return 16
        else:
            return 21
    elif lis[0] == 2:
        if lis[1] == 1:
            return 2
        elif lis[1] == 2:
            return 7
        elif lis[1] == 3:
            return 12
        elif lis[1] == 4:
            return 17
        else:
            return 22
    elif lis[0] == 3:
        if lis[1] == 1:
            return 3
        elif lis[1] == 2:
            return 8
        elif lis[1] == 3:
            return 13
        elif lis[1] == 4:
            return 18
        else:
            return 23
    elif lis[0] == 4:
        if lis[1] == 1:
            return 4
        elif lis[1] == 2:
            return 9
        elif lis[1] == 3:
            return 14
        elif lis[1] == 4:
            return 19
        else:
            return 24
    else:
        if lis[1] == 1:
            return 5
        elif lis[1] == 2:
            return 10
        elif lis[1] == 3:
            return 15
        elif lis[1] == 4:
            return 20
        else:
            return 25
    
```

```python
data_grid = data
data_grid['grid'] = 0
l = len(data_grid.index)
print l
data_grid.head()
```

```python
get_grid(give_cell(data_grid.positions[0][0]))
```

```python
for i in range(l):
    temp = get_grid(give_cell(data_grid.positions[i][0]))
    data_grid.grid[i] = temp
```
