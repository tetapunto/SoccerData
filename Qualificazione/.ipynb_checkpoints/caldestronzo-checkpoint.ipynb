ESERCIZIO:
Calcolare la posizione media di ogni calciatore che abbia giocato almeno 45 minuti. Calcolare inoltre la distanza quadratica media di ogni giocatore rispetto alla sua posizione media. Calcolare, per ogni giocatore, il tempo medio che passa tra due eventi del giocatore (in secondi). 

```python
# add 45 * 60 minutes for events in the second half
matches.eventSec = matches.eventSec + matches.matchPeriod.apply(lambda s: 0 if s[0] == '1' else 45 * 60)
no_players = [k for k, v in (matches.groupby('playerId').agg({'teamId': 'nunique'}).teamId < 2).to_dict().items() 
              if not v]
matches.drop(no_players, inplace=True)
```

```python
players_45min = matches\
.groupby('playerId', as_index=False).agg({"eventSec": [np.max, np.min]})\
.assign(playing_time=lambda  x: (x[('eventSec', 'amax')] - x[('eventSec', 'amin')]) / 60)\
.loc[:, ['playerId', 'playing_time']]
```

sommatoria, **per ogni posizione p del calciatore**, della distanza al quadrato tra la posizione p e la posizione media, fratto **il numero totale di eventi del calciatore** nella partita

```python
player_45min_ids = players_45min[players_45min.playing_time >= 45].playerId.values
matches = matches.loc[matches.playerId.isin(player_45min_ids)]

def extract_position(series_pos):
    ys = np.array([d['y'] for line in series_pos.values for d in line])
    xs = np.array([d['x'] for line in series_pos.values for d in line])
    y_mean = ys.mean()
    x_mean = xs.mean()
    dist = np.power(ys - y_mean, 2) + np.power(xs - x_mean, 2)
    return pd.DataFrame([[x_mean, y_mean, dist.sum() / len(series_pos)]],
                       columns=['posizione_media_x', 'posizione_media_y', 'distanza_quadratica_media'])

result1 = matches.groupby('playerId').positions.apply(extract_position)
result1.index = result1.index.droplevel(1)
```

```python
def extract_time_delta(series_time):
    return np.mean(series_time - series_time.shift(1))

result2 = matches.groupby('playerId').eventSec.apply(extract_time_delta).rename('tempo_medio_tra_eventi')
```

```python
matches[['playerId', 'teamId']].drop_duplicates('playerId')\
.merge(players, how='left', on='playerId')\
.merge(pd.concat((result1, result2), axis=1).reset_index(), how='left', on='playerId')\
.rename(columns={'playerId': 'identificativo_calciatore',
                 'name': 'nome_calciatore',
                 'teamId': 'squadra_calciatore'})\
.loc[:, ['identificativo_calciatore', 'nome_calciatore', 'squadra_calciatore', 'posizione_media_x',
         'posizione_media_y', 'distanza_quadratica_media', 'tempo_medio_tra_eventi']]\
.to_csv('problema_1.csv', index=False)
     
```

ESERCIZIO (a):
Dividere il campo di gioco in una griglia 5x5 (ossia in 25 celle di uguale dimensione). Per ogni cella e per ogni squadra calcolare:
i. Frequenza degli eventi effettuati (numero di eventi in quella cella sul totale degli eventi della squadra)
ii. Frequenza di passaggi accurati effettuati (numero di passaggi accurati in quella cella sul totale dei passaggi accurati)
iii. Frequenza di tiri effettuati 
iv. Frequenza di falli subiti 

```python
def transform_tags(l):
    if len(l) == 0:
        return np.nan
    elif 1801 in list(l[0].values()):
        return 'accurate'
    return 'non_accurate'

matches.tags = matches.tags.apply(transform_tags)
```

```python
def extract_positions(pos_series):
    pos_list = pos_series.tolist()[0]
    try:
        y_end = pos_list[1]['y']
        x_end = pos_list[1]['x']
    except:
        y_end = np.nan
        x_end = np.nan
    y_start = pos_list[0]['y']
    x_start = pos_list[0]['x']
    return pd.DataFrame([[x_start, y_start, x_end, y_end]], columns=['x_start', 'y_start', 'x_end', 'y_end'])

positions = matches.groupby('id').positions.apply(extract_positions)
positions.index = positions.index.droplevel(1)
matches = matches.merge(positions.reset_index(), on='id')
```

```python
def cell_number(pos):
    return (pos / 20.01).astype('int')

cells = matches.set_index('id').iloc[:, -4:].apply(cell_number)

matches_with_cell_numbers = pd.concat((cells.iloc[:, :2].rename(columns={'x_start': 'numero_cella_x',
                                             'y_start': 'numero_cella_y'}), 
           cells.iloc[:, -2:].rename(columns={'x_end': 'numero_cella_x',
                                             'y_end': 'numero_cella_y'})
          )).reset_index()\
.merge(matches, on='id')


r1 = matches_with_cell_numbers.groupby(['teamId', 'numero_cella_x', 'numero_cella_y']).agg({'id': 'nunique'})\
.rename(columns={'id': 'frequenza_eventi'})

r2 = matches_with_cell_numbers[(matches_with_cell_numbers.tags == 'accurate') &
                               (matches_with_cell_numbers.eventName == 'Pass')]\
.groupby(['teamId', 'numero_cella_x', 'numero_cella_y']).agg({'id': 'nunique'})\
.rename(columns={'id': 'frequenza_passaggi_accurati'})
```

```python
total = pd.concat((r1,
           r2,  
           matches_with_cell_numbers[matches_with_cell_numbers.eventName == "Shot"]\
           .groupby(['teamId', 'numero_cella_x', 'numero_cella_y']).agg({'id': 'nunique'})\
           .rename(columns={'id': 'frequenza_tiri'}),
           
           matches_with_cell_numbers[matches_with_cell_numbers.eventName == 'Foul']\
           .groupby(['teamId', 'numero_cella_x', 'numero_cella_y']).agg({'id': 'nunique'})\
           .rename(columns={'id': 'frequenza_falli'})
          ), axis=1).fillna(0).reset_index().rename(columns={'teamId': 'nome_squadra'})

for c in total.columns[-4:]:
    for s in total.nome_squadra.unique():
        total.loc[(total.nome_squadra == s), c] = \
        (total.loc[(total.nome_squadra == s), c] / total.loc[(total.nome_squadra == s), c].sum()).round(2)
        
total.to_csv('problema_3_a.csv', index=False)
```

## problema B

```python
from scipy.spatial.distance import cosine

total = pd.read_csv('problema_3_a.csv')
```

```python
similarities = []

for c in total.columns[-4:]:
    res = cosine(total.loc[total.nome_squadra == 'Croatia', c].values, 
                 total.loc[total.nome_squadra == 'France', c].values)
    similarities.append(res)
```

```python
pd.DataFrame(list(zip(range(1,5), similarities)), columns=['tipo_griglia','similarity']).to_csv('problema_3_b.csv',
                                                                                               index=False)
```

## problema C
Calcolare la squadra “dominante” in ogni cella delle griglia costruita secondo il punto (a)-i. Una squadra è “dominante” in una cella se il suo valore di frequenza è maggiore di quello della squadra avversaria

```python
total = pd.read_csv('problema_3_a.csv').set_index(['numero_cella_x', 'numero_cella_y'])
origin =  (total.loc[total.nome_squadra == 'France', 'frequenza_eventi'] - total.loc[total.nome_squadra == 'Croatia',
                                                                                     'frequenza_eventi'])

for c in total.columns[-3:]:
    origin = pd.concat((origin,
               total.loc[total.nome_squadra == 'France', c] - total.loc[total.nome_squadra == 'Croatia', c]
              ), axis=1)
```

```python
origin
```

```python
origin.columns = range(1, 5)
origin = origin.stack().reset_index().rename(columns={'level_2': 'tipo_griglia', 0:'differenza_di_frequenza'})
origin['squadra_dominante'] = origin.differenza_di_frequenza.apply(lambda b: 'France' if b >= 0 else 'Croatia')
origin.differenza_di_frequenza = origin.differenza_di_frequenza.abs()

origin.loc[:, ['tipo_griglia', 'numero_cella_x', 'numero_cella_y', 'squadra_dominante', 'differenza_di_frequenza']]\
.to_csv('problema_3_c.csv', index=False)
```
